## Relationship between wards where BRT use is above 15% and attitudes to foreigners/other population groups

Here I will explore the wards where BRT use is above 15%. In particular I am interested in understanding what the racial breakdown is, and what the relationship is between wards where there are frequent BRT users and their attitude

```python
# Importing all packages that will be necessary for the data exploration 

# Importing the pandas library 
import pandas as pd

#For creating or removing a directory folder
import os

#To allow me to work with numerical data
import numpy as np

# for spatial data
import geopandas as gpd

# for plotting
import matplotlib.pyplot as plt

# for network analysis
import networkx as nx

# for street network analysis
import osmnx as ox

# for basemaps
import contextily as ctx
```

Importing edited csv file with wards that have over 15% BRT use

```python
BRT15=pd.read_csv('BRT15%.csv')
wards_shp=gpd.read_file("MDB_Wards_2020.shp")
```

Renaming the columns so that the ward and WardID columns match up

```python
BRT15=BRT15.rename(columns={"ward":"WardID"})
```

```python
BRT15=BRT15.rename(columns={"Q9_10_BW_trust":"Blacks and whites will never really trust each other"})
```

```python
wards_shp['WardID'] = wards_shp['WardID'].astype(int)

wards_shp.dtypes
```

```python
BRT15.head()
```

Merging shapefile and csv, on WardID column and viewing the value counts

```python

BRT15_mer = wards_shp.merge(BRT15, on='WardID')
```

```python
BRT15_mer['Blacks and whites will never really trust each other'].value_counts()
```

Now, I'd like to plot this, which I do in a bar chart

```python
BWtrust = BRT15_mer['Blacks and whites will never really trust each other'].value_counts()

BWtrust = BWtrust.reset_index()

BWtrust.head()


BWtrust.plot.bar(figsize=(12,6),
                            x='index',
                            y='Blacks and whites will never really trust each other',
                          color=['green'],
                            title="Level of trust between Black and White people in Wards where BRT use is over 15%")
```

I would now like to map this, although I later realise that all the respondents are in one ward therefore mapping it doesn't make much sense. However, by going through the exercise, I do see what the mean response is across the ward, which is useful. 

```python
BRT15_mer=BRT15_mer.rename(columns={"Blacks and whites will never really trust each other":"Q9_10_BW_trust"})
```

```python
BRT15_mer.loc[BRT15_mer['Q9_10_BW_trust']=='Strongly disagree','BW_trust']=4
BRT15_mer.loc[BRT15_mer['Q9_10_BW_trust']=='Disagree','BW_trust']=3
BRT15_mer.loc[BRT15_mer['Q9_10_BW_trust']=='Neither agree nor disagree','BW_trust']=2
BRT15_mer.loc[BRT15_mer['Q9_10_BW_trust']=='Agree','BW_trust']=1
BRT15_mer.loc[BRT15_mer['Q9_10_BW_trust']=='Strongly agree','BW_trust']=0
BRT15_mer [['Q9_10_BW_trust','BW_trust']]
```

```python
BRT15_mer.head()
```

```python
#create ward level dataset with ward average bw_trust variable
BRT_BW_trust = BRT15_mer.groupby("WardID").mean('BW_trust') [["BW_trust"]]

# check ward level trust var
BRT_BW_trust.head()

# merge ward level trust var back to ward level shape data (again)

BRT_BW_trust = BRT15_mer.merge(BRT_BW_trust, on='WardID')

BRT_BW_trust.head()
```

## Plot the ward level data and realise, that this doesn't add anything visually, as it is all in the same ward.

## Bar chart, attitudes towards foreigners

Here I create a bar chart of the attitudes towards foreigners, in the same way I did above for the BW trust variable

```python
foreigners_att = BRT15_mer['Q9_12_foreigners'].value_counts()

foreigners_att = foreigners_att.reset_index()

foreigners_att.head()


```

```python
foreigners_att.plot.barh(figsize=(12,6),
                            x='index',
                            y='Q9_12_foreigners',
                          color=['green'],
                            title="Attitudes towards foreigners")
```

Here I can see that the majority of respondents thing that foreigners are ok, but need legal permission from the government. This is interesting given the amount of informal traders and the number of xenophobic attacks against them, as these traders are in South Africa illegally.
